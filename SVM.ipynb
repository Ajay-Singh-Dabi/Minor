```python
collections_c = []
collections_gamma = []
accuracy = []
```

```python
for i in range(1,6):
  training_dataset = pd.read_csv(f"./pima-5-fold/pima-5-{i}tra.csv")
  testing_dataset = pd.read_csv(f"./pima-5-fold/pima-5-{i}tst.csv")

  #split dataset in features and target variable:
  x = training_dataset.iloc[:,:8] #feature
  y = training_dataset['Result'] # target

  from sklearn.model_selection import train_test_split
  x_train, y_train = x,y

  x_test = testing_dataset.iloc[:,:8] #feature
  y_test = testing_dataset['Result'] # target

  from sklearn.metrics import classification_report, confusion_matrix
  from sklearn.svm import SVC

  model = SVC()
  model.fit(x_train, y_train)

  prediction = model.predict(x_test)
  print(classification_report(y_test, prediction))

  from sklearn.model_selection import GridSearchCV
  import numpy as np

  c_list = []
  for i in np.arange(-18,18,2):
    c_list.append(float(2)**i)

  gamma_list = []
  for i in np.arange(-18,18,2):
    gamma_list.append(float(2)**i)

  param_grid = {'C' : c_list,
                'gamma' : gamma_list,
                'kernel' : ['rbf']}
  clf = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)

  clf.fit(x_train, y_train)

  best_param = clf.best_params_
  collections_c.append(best_param)
  best_estimator = clf.best_estimator_
  collections_gamma.append(best_estimator)
  print(best_param)
  print(best_estimator)

  grid_prediction  = clf.predict(x_test)
  print(classification_report(y_test, grid_prediction))
  from sklearn.metrics import accuracy_score
  accuracy.append(accuracy_score(y_test, clf.predict(x_test)))
  print(accuracy_score(y_test, clf.predict(x_test)))
```
